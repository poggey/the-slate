```python
import pandas as pd
df = pd.read_csv("../processed/films_raw.csv")

df.shape
df.info()
df.isna().sum()
df.describe()
df.sort_values("budget_usd").head(10)
```

```python
## check for duplicated and remove short films

df.duplicated().sum()
df = df.drop_duplicates()
df = df[df["runtime_min"] >= 40]
df.head()
```

```python
## dont want to delete rows so flag rows with missing data
import numpy as np

df["flag_no_budget"]   = df["budget_usd"].isna() | (df["budget_usd"] == 0)
df["flag_no_revenue"]  = df["worldwide_gross_usd"].isna() | (df["worldwide_gross_usd"] == 0)
df["flag_still_in_release"] = df["release_year"] >= 2026   # current year only — 2025 grosses are final by mid-2026

## distribution_only, hand-coded from the A24 filmography as planned — the cached
## Wikipedia page marks every acquisition in its Notes column ("U.S. distribution
## only", "North American co-distribution with CMC Pictures only", ...); films
## without that note are A24 (co-)productions. TMDB's production_companies is too
## unreliable for this (it omits A24 on Hereditary, Lady Bird, Midsommar).
import io, re

def norm_title(t):
    return re.sub(r"[^a-z0-9]+", " ", str(t).lower()).strip()

def is_dist_only_note(note):
    ## "distribution ... only" within one sentence, so plain "distribution only"
    ## and "co-distribution with CMC Pictures only" (Ne Zha 2) both match
    return bool(re.search(r"distribution\b[^.]*\bonly\b", str(note).lower()))

wiki_tables = pd.read_html(io.StringIO(open("../raw/wikipedia/a24_filmography.html").read()))
dist_only_titles = set()
for t in wiki_tables:
    t.columns = [" ".join(map(str, c)) if isinstance(c, tuple) else str(c) for c in t.columns]
    title_col = next((c for c in t.columns if "title" in c.lower()), None)
    notes_col = next((c for c in t.columns if "note" in c.lower()), None)
    if title_col is None or notes_col is None:
        continue
    for _, r in t.iterrows():
        if is_dist_only_note(r[notes_col]):
            dist_only_titles.add(norm_title(re.sub(r"\[[^\]]*\]", "", str(r[title_col]))))

df["flag_distribution_only"] = df["title"].map(norm_title).isin(dist_only_titles)

df[[c for c in df.columns if c.startswith("flag_")]].sum()
```

```python
## see if theres any nonsense budget data

df.sort_values("budget_usd", na_position="first")[
    ["title", "release_year", "budget_usd", "worldwide_gross_usd"]
].head(20)
```

```python
df["in_core"] = ~(df["flag_no_budget"] | df["flag_no_revenue"])
## §1.3: distribution-only deals carry no production risk and their budgets/grosses
## aren't A24 economics (Ne Zha 2 alone adds ~$1.2B of phantom profit) — out of core,
## kept as rows for the §8 robustness rerun
df["in_core"] &= ~df["flag_distribution_only"]
print(f"Core sample: {df['in_core'].sum()} of {len(df)} films have both budget and revenue")
```

```python
df["genres"].head(20)
def first_genre(val):
    if pd.isna(val):
        return None
    return val.split("|")[0]   # take the first genre before the pipe

df["primary_genre"] = df["genres"].apply(first_genre)
# TMDB's genre order is arbitrary, so first-position coding hid 9 horror films
# (Heretic, Bodies Bodies Bodies, It Comes at Night...) inside Drama/Comedy.
# H3 hinges on horror, so horror takes priority over position:
df.loc[df["genres"].str.contains("Horror", na=False), "primary_genre"] = "Horror"
df["primary_genre"].value_counts(dropna=False)
```

```python
df[df["primary_genre"] == "Horror"]["title"]        # does this list look complete?
# and check what's hiding horror in a later position:
df[df["genres"].str.contains("Horror") & (df["primary_genre"] != "Horror")][["title", "genres"]]
```

```python
keep = ["Drama", "Comedy", "Horror", "Thriller", "Crime"]   # adjust to actual counts
df["primary_genre"] = df["primary_genre"].where(df["primary_genre"].isin(keep), "Other")
df["primary_genre"].value_counts()
```
